# Lab - Linearidade TF

Nesta tarefa voce deve verificar a linearidade da transformada de Fourier ($\mathcal{F}$), usando a função fft().<br>
Considere os seguintes sinais, para $n=0...99$:<br>
$x_1[n]=sin(w_1 n)\qquad w_1={2\pi}/{N}$<br>
$x_2[n]=sin(w_2 n)\qquad w_2={4\times2\pi}/{N}$<br>
e as constantes $a=4$ e $b=2$<br>
Mostre que $a\, \mathcal{F}\{ x_1[n]\} + b\, \mathcal{F}\{x_2[n]\} = \mathcal{F}\{a x_1[n] + b x_2[n] \}$

*Processamento Digital de Sinais - UFPR - Prof. Eduardo Parente Ribeiro*

```python
from pylab import *
```

Primeiramente defina os sinais x1, x2 e constantes a, b. Plot os sinais. Plote a transformada de Fourier de cada sinal com a respectiva ponderação: $a\, \mathcal{F}\{ x_1[n]\} + b\, \mathcal{F}\{x_2[n]\}$

```python
N=100
W=2*pi/N
n=arange(N)
w1=1*W
w2=4*W
x1=sin(w1*n)
x2=sin(w2*n)
a=4
b=2
X1=fft(x1)/N
X2=fft(x2)/N

figure(figsize=(15,4))
subplot(1,2,1)
plot(n,x1,label='x1[n]')
plot(n,x2,label='x2[n]')
legend()
subplot(1,2,2)
stem(n,a*abs(X1),linefmt='C0-',markerfmt='C0o', basefmt='C0-', label='a.|F{x1}|')
stem(n,b*abs(X2),linefmt='C1-',markerfmt='C1o', basefmt='C1-', label='b.|F{x2}|')
legend()
show()
```

Agora obtenha a a soma ponderada, plotando o sinal obtido e sua transformada de Fourier: $x[n]=a x_1[n] + b x_2[n]\, , \,\mathcal{F}\{x[n] \}$. <br>
Plote os resultados, confirmando que que $a\, \mathcal{F}\{ x_1[n]\} + b\, \mathcal{F}\{x_2[n]\} = \mathcal{F}\{a x_1[n] + b x_2[n] \}$

```python
# Soma ponderada dos sinais
x = a * x1 + b * x2
X = fft(x) / N

# Plot do sinal resultante plot da sua transformada de Fourier
figure(figsize=(15, 4))
subplot(1, 2, 1)
plot(n, x, label='a x1[n] + b x2[n]')
legend()

# Plot TF
subplot(1, 2, 2)
stem(n, abs(X), linefmt='C2-', markerfmt='C2o', basefmt='C2-', label='|F{ax1 + bx2}|')
legend()
show()
```
