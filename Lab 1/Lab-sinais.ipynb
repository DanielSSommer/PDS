### a) impulso unit√°rio, para valores de n (abscissa) entre -10 e 10 (stem), inclusive.

```python
from pylab import *

# Valores de n de -10 a 10
n = arange(-10, 11)

# Gerando o sinal de impulso unit√°rio
sinal_impulso = where(n == 0, 1, 0)

# Usando o comando stem para visualiza√ß√£o
stem(n, sinal_impulso)
title('a) Impulso Unit√°rio')
xlabel('n')
ylabel('Amplitude')
show()
```

### b) degrau unit√°rio, para n entre -10 e 10 (stem), inclusive.

### d) senoide com 4 ciclos, com 90 valores, para n=0 a 89 (stem).

```python
# n√∫mero valores
valores = 90

# frequ√™ncia da senoite (4 ciclos com 90 valores)
frequencia = 4/valores

# valores de n de 0 a 89:
n = arange(0, valores)

# senoide:
senoide = sin(2*pi*frequencia*n)

# Gr√°fico:
stem(n, senoide)
title('d) senoide com 4 ciclos, com 90 valores')
xlabel('n')
ylabel('Amplitude')
show()
```

### e) cosenoide com 5 ciclos, com 110 valores, para n=0 a 109 (stem). 

```python
# n√∫mero valores
valores = 110

# frequ√™ncia da senoite (4 ciclos com 90 valores)
frequencia = 5/valores

# valores de n de 0 a 89:
n = arange(0, valores)

# senoide:
cosenoide = cos(2*pi*frequencia*n)

# Gr√°fico:
stem(n, cosenoide)
title('d) cosenoide com 5 ciclos, com 110 valores')
xlabel('n')
ylabel('Amplitude')
show()
```

### f) exponencial complexa com 5 ciclos, para n=0 a 69. Parte real (stem) e parte imagin√°ria (plot). Usar subplot para plotar em dois graficos separados: subplot(2,1,1) para o primeiro e subplot(2,1,2) para do segundo.

```python
# n√∫mero valores
valores = 70

# frequ√™ncia da senoite (4 ciclos com 90 valores)
frequencia = 5/valores

# valores de n de 0 a 89:
n = arange(0, valores)

# senoide:
exp_complexa = exp(1j*2*pi*frequencia*n)

# Parte Real:
parte_real = real(exp_complexa)
parte_imaginaria = imag(exp_complexa)

# Subplots:
subplot(2, 1, 1)
stem(n, parte_real)
title('Parte Real da Exponencial Complexa')
xlabel('n')
ylabel('Amplitude')

subplot(2, 1, 2)
plot(n, parte_imaginaria, color='g')
title('Parte Imagin√°ria da Exponencial Complexa')
xlabel('n')
ylabel('Amplitude')

subplots_adjust(hspace=0.5)

show()
```

### g) senoide com frequ√™ncia w=0,2 rd ( sin(w*n) ) , para n de 0 a 99 (plot).

```python
# n√∫mero valores
valores = 100

# frequ√™ncia da senoite (4 ciclos com 90 valores)
freq_angular = 0.2

# valores de n de 0 a 89:
n = arange(0, valores)

# senoide:
senoide = sin(freq_angular*n)

# Gr√°fico:
plot(n, senoide)
title('g) Senoide com Frequ√™ncia Angular $\\omega = 0.2$ rad')
xlabel('n')
ylabel('Amplitude')
show()
```

### h) senoide de 60 Hz, (amostragem = 800 amostras/s), com 70 valores, para n de 0 a 69 (plot).

```python
# n√∫mero valores:
valores = 70

# gerando 70 valores
n = arange(0, valores)

# frequ√™ncia:
frequencia = 60

# per√≠odo:
periodo = 1/frequencia

# Intervalo entre as amostras:
intervalo_amostras = periodo/valores

# tempo de amostragem:
tempo_amostragem = n*intervalo_amostras

# senoide:
senoide = sin(2 * pi * frequencia * n / 800)

# Gr√°fico:
plot(n, senoide)
title('h) Senoide de 60 Hz com 800 amostras/s')
xlabel('n')
ylabel('Amplitude')
show()
```

### l) exponencial decrescente para n = 0 a 60 (stem).

```python
# n√∫mero valores
valores = 61

# decaimento
decaimento = -0.1

# valores de n de 0 a 60:
n = arange(0, valores)

# exponencial decrescente:
exp_decrescente = exp(decaimento * n)

# Gr√°fico:
stem(n, exp_decrescente)
title('I) Exponencial Decrescente')
xlabel('n')
ylabel('Amplitude')
show()
```

### m) exponencial oscilante decrescente $x[n]=z^n$, para n=0 a 90, parte real (stem).

```python
# n√∫mero valores
valores = 91

# Complexo z:
z = complex(0.9,0.4)

# valores de n de 0 a 90:
n = arange(0, valores)

# Exponencial oscilante decrescente
exp_oscilante_decrescente = real(z ** n)

# Gr√°fico:
stem(n, exp_oscilante_decrescente)
title('m) Exponencial Oscilante Decrescente (Parte Real)')
xlabel('n')
ylabel('Amplitude')
show()
```

### n) exponencial oscilante crescente $x[n]=z^n$, para n=0 a 80, parte imaginaria (stem).

```python
# n√∫mero valores
valores = 81

# Complexo z:
z = complex(1,0.3)

# valores de n de 0 a 90:
n = arange(0, valores)

# Exponencial oscilante crescente
exp_oscilante_crescente = imag(z ** n)

# Gr√°fico:
stem(n, exp_oscilante_crescente)
title('n) Exponencial Oscilante Crescente (Parte Imagin√°ria)')
xlabel('n')
ylabel('Amplitude')
show()
```
