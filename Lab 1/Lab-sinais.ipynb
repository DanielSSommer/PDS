### a) impulso unitário, para valores de n (abscissa) entre -10 e 10 (stem), inclusive.

```python
from pylab import *

# Valores de n de -10 a 10
n = arange(-10, 11)

# Gerando o sinal de impulso unitário
sinal_impulso = where(n == 0, 1, 0)

# Usando o comando stem para visualização
stem(n, sinal_impulso)
title('a) Impulso Unitário')
xlabel('n')
ylabel('Amplitude')
show()
```

### b) degrau unitário, para n entre -10 e 10 (stem), inclusive.


```python
# criando o intervalo:
n = arange(-10,11)

step_sinal=  where(n>=0,1,0)
stem(n, step_sinal)
title('b) degrau unitário')
xlabel('n')
ylabel('Amplitude')
show()
```

### c) ruido branco gaussiano (média zero, variância unitária) com 40 valores, para n=0 a 39 (stem).

```python
# para 40 valores:
valores = 40

# ruido branco gaussiano com média zeroe variância unitária:
ruido_branco = randn(valores)

# gerando valores de 0 a 39:
n = arange(0, valores)

# visualização:
stem(n, ruido_branco)
title('c) Ruído Branco Gaussiano (média zero, variância unitária)')
xlabel('n')
ylabel('Amplitude')
show()

```

### d) senoide com 4 ciclos, com 90 valores, para n=0 a 89 (stem).

```python
# número valores
valores = 90

# frequência da senoite (4 ciclos com 90 valores)
frequencia = 4/valores

# valores de n de 0 a 89:
n = arange(0, valores)

# senoide:
senoide = sin(2*pi*frequencia*n)

# Gráfico:
stem(n, senoide)
title('d) senoide com 4 ciclos, com 90 valores')
xlabel('n')
ylabel('Amplitude')
show()
```

### e) cosenoide com 5 ciclos, com 110 valores, para n=0 a 109 (stem). 

```python
# número valores
valores = 110

# frequência da senoite (4 ciclos com 90 valores)
frequencia = 5/valores

# valores de n de 0 a 89:
n = arange(0, valores)

# senoide:
cosenoide = cos(2*pi*frequencia*n)

# Gráfico:
stem(n, cosenoide)
title('d) cosenoide com 5 ciclos, com 110 valores')
xlabel('n')
ylabel('Amplitude')
show()
```

### f) exponencial complexa com 5 ciclos, para n=0 a 69. Parte real (stem) e parte imaginária (plot). Usar subplot para plotar em dois graficos separados: subplot(2,1,1) para o primeiro e subplot(2,1,2) para do segundo.

```python
# número valores
valores = 70

# frequência da senoite (4 ciclos com 90 valores)
frequencia = 5/valores

# valores de n de 0 a 89:
n = arange(0, valores)

# senoide:
exp_complexa = exp(1j*2*pi*frequencia*n)

# Parte Real:
parte_real = real(exp_complexa)
parte_imaginaria = imag(exp_complexa)

# Subplots:
subplot(2, 1, 1)
stem(n, parte_real)
title('Parte Real da Exponencial Complexa')
xlabel('n')
ylabel('Amplitude')

subplot(2, 1, 2)
plot(n, parte_imaginaria, color='g')
title('Parte Imaginária da Exponencial Complexa')
xlabel('n')
ylabel('Amplitude')

subplots_adjust(hspace=0.5)

show()
```

### g) senoide com frequência w=0,2 rd ( sin(w*n) ) , para n de 0 a 99 (plot).

```python
# número valores
valores = 100

# frequência da senoite (4 ciclos com 90 valores)
freq_angular = 0.2

# valores de n de 0 a 89:
n = arange(0, valores)

# senoide:
senoide = sin(freq_angular*n)

# Gráfico:
plot(n, senoide)
title('g) Senoide com Frequência Angular $\\omega = 0.2$ rad')
xlabel('n')
ylabel('Amplitude')
show()
```

### h) senoide de 60 Hz, (amostragem = 800 amostras/s), com 70 valores, para n de 0 a 69 (plot).

```python
# número valores:
valores = 70

# gerando 70 valores
n = arange(0, valores)

# frequência:
frequencia = 60

# período:
periodo = 1/frequencia

# Intervalo entre as amostras:
intervalo_amostras = periodo/valores

# tempo de amostragem:
tempo_amostragem = n*intervalo_amostras

# senoide:
senoide = sin(2 * pi * frequencia * n / 800)

# Gráfico:
plot(n, senoide)
title('h) Senoide de 60 Hz com 800 amostras/s')
xlabel('n')
ylabel('Amplitude')
show()
```

### l) exponencial decrescente para n = 0 a 60 (stem).

```python
# número valores
valores = 61

# decaimento
decaimento = -0.1

# valores de n de 0 a 60:
n = arange(0, valores)

# exponencial decrescente:
exp_decrescente = exp(decaimento * n)

# Gráfico:
stem(n, exp_decrescente)
title('I) Exponencial Decrescente')
xlabel('n')
ylabel('Amplitude')
show()
```

### m) exponencial oscilante decrescente $x[n]=z^n$, para n=0 a 90, parte real (stem).

```python
# número valores
valores = 91

# Complexo z:
z = complex(0.9,0.4)

# valores de n de 0 a 90:
n = arange(0, valores)

# Exponencial oscilante decrescente
exp_oscilante_decrescente = real(z ** n)

# Gráfico:
stem(n, exp_oscilante_decrescente)
title('m) Exponencial Oscilante Decrescente (Parte Real)')
xlabel('n')
ylabel('Amplitude')
show()
```

### n) exponencial oscilante crescente $x[n]=z^n$, para n=0 a 80, parte imaginaria (stem).

```python
# número valores
valores = 81

# Complexo z:
z = complex(1,0.3)

# valores de n de 0 a 90:
n = arange(0, valores)

# Exponencial oscilante crescente
exp_oscilante_crescente = imag(z ** n)

# Gráfico:
stem(n, exp_oscilante_crescente)
title('n) Exponencial Oscilante Crescente (Parte Imaginária)')
xlabel('n')
ylabel('Amplitude')
show()
```
